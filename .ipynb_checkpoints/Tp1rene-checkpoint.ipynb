```python
%pylab inline
import timeit
import time
import numpy as np 
from scipy import signal as sig
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft, fftshift
from __future__ import division

#principio#

Fs=10000   #frecuencia de muetreo d 10 khz#
N=200      #numero de muestras#
t=np.linspace(0,N/Fs,num=2*N)  #crea la variable de tiempo desde 0, con intervalos de 1/Fs, hasta 2*N #
F1=2000
F2=1000

# punto 1, calculo de la trasferencia#


zeros=np.zeros(100)
ones=np.ones(1)
dirac=np.concatenate((zeros,ones,zeros),axis=0) # crea una delta de dirac de cien ceros un uno y cien ceros

#punto 2  cuadrada y triangular atraez del movingaverage#

cuad=sig.square(2*np.pi*F1*t)

trian=sig.sawtooth(2*np.pi*F2*t,0.5)

MVSIZE=3 #ventana movingaverage#

def mov_avg (signal, MVSIZE):
    avg = np.zeros(len(signal))
    for n in range(len(avg)):
        sum = 0
        d = 0
        for k in range (-(MVSIZE), MVSIZE):
            if ((n-k) > 0) and ((n-k) < (len(signal))):
                sum = sum + signal[(n-k)]
                d = d+1
        avg[n] = sum/d
    return avg






transfe=mov_avg(dirac,MVSIZE)   #movingaverage dirac

cuad_fil=mov_avg(cuad,MVSIZE)   #movingaverage a la cuadrada

trian_fil=mov_avg(trian,MVSIZE)  #movingaverage a la triangular





plt.plot(abs(fft(transfe))) #muetra la transferencia despues de pasarle una dirac al muving average
plt.title('transferencia movingaverage punto 1')
plt.figure()

plt.plot(t,cuad_fil)
plt.title('cuadrada punto 2')
plt.figure()

plt.title('triangular punto 2')
plt.plot(t,trian_fil)




```


#Conclusion punto 1 y 2


Al hacer pasar una dirac por el movingaverage, se obtendra un pulso cuadrado unico
cuyo ancho dependera de cuan grande sea la ventana MVSIZE. El espectro de este pulso es la que corresponde a una señal
SING como se ve en la primera grafica.
cuando hacemos pasar la triangular y la cuadrada por el movingaverage, vemos como se van redondeandolas puntas pel recorte de las altas frecuncias con el aumento de la ventana MVSIZE.







```python


#Punto 3. medicion de los tiempos para la convolucion en variable int y float.


t1 = np.arange(0, 1024/20000, 1/20000)
s1= 10*sin(2*np.pi*1000*t1)+1 
s2 = 10*sin(2*np.pi*1200*t1)

def convolu(f1, f2):
    
    
    voltf1=f1[::-1]                                           #voltea la señal f1 para hacer la convolucion
                                                              
    cl=np.zeros(f1.size+f2.size-1)                            # crea un arreglo de ceros de tamaño f1  mas f2  menos uno 
    
    cceros=np.insert(f2,0,np.zeros(f1.size-1))                #coloco ceros al principio para hacer la convolucion
    
    cceros=np.insert(cceros,cceros.size,np.zeros(f1.size-1))  #coloco cantidad de ceros necesaria al final
    
                                                              #ciclos for de n y k para recorrer la convolucion
    
    for n in range(cl.size):
        for k in range(f1.size):
            ind= n+k
            cl[n]=cl[n]+voltf1[k]*cceros[ind]
    return cl    
    

def convolu_int(f1, f2):
    
    
    voltf1=f1[::-1]                                #voltea la señal f1 modo int para hacer la convolucion
                                                              
    cl=np.zeros(f1.size+f2.size-1)                 # crea un arreglo de ceros de tamaño f1  mas f2  menos uno 
    
    cceros=np.insert(f2,0,np.zeros(f1.size-1))        #coloco ceros al principio para hacer la convolucion
    
    cceros=np.insert(cceros,cceros.size,np.zeros(f1.size-1))      #coloco cantidad de ceros necesaria al final
    
                                                                  #ciclos for de n y k para recorrer la convolucion
    
    for n in range(cl.size):
        for k in range(f1.size):
            ind= n+k
            cl[n]=int(cl[n])+int(voltf1[k])*int(cceros[ind])     #trabaja con la parte entera de los arreglos y hace la 
    return cl                                                    #multiplicacion   

print("convolucion con numpy")
print(convolve(s1,s2)) 
print("convolocion algoritmo float")
print(convolu(s1,s2))
print("convolucion algoritmo int")
print(convolu_int(s1,s2))
    
        

start1 = time.time()         #medicion de tiempo que tarda la convolucion en float
conv_r = np.convolve(s1,s2)
end1 = time.time()
print("tiempo que tarda la onvolucion en numpy")
print (end1-start1)


start2 = time.time()         #medicion de tiempo que tarda la convolucion en float
conv_r = convolu(s1,s2)
end2 = time.time()
print("tiempo que tarda la convolucion en float")
print (end2-start2)
   
start3=time.time() 
conv_r=convolu_int(s1,s2)
end3=time.time()
print("tiempo que tarda la convolucion en int")
print(end3-start3)



```
